```python
def get_list_valid_names(df, filter_colum, get_colum,num_camera ,arg_filter = 'S'):
    def get_name(string, num_cam):
        return f"test2_cam{num_cam}_{string[1:]}.MP4"
    _df1 = df[df[filter_colum].str.contains(arg_filter)]
    return _df1[get_colum].apply(get_name,num_cam=num_camera).values.tolist()
```

```python
list1 = get_list_valid_names(df_clasificacion_video , "CAM 1" , "Gesto/Cámara","1")
list2 = get_list_valid_names(df_clasificacion_video , "CAM 2" , "Gesto/Cámara","2")
with open("new_input", "w") as file:
    [file.write(l+"\n") for l in sorted(list1+list2)]
```
